### Create Artist Lookup Dictionary

Next, I'm going to create a lookup dictionary for artists with their index. This will make it very fast to look up whether an artist already has an index or not.

The rationale for this is as follows. We have 1769 listed artists with indexes already assigned. We don't want to reassign unique indexes; these may be in use somewhere already, so we want to retain unique indexes for all artists that already have one assigned. However, it is clear we have a lot more artists in the data than have these inexes assigned. So as we look over the unique list of all artists, we are going to want to look up whether it has an index, and if not assign it a new one. To make this an efficient and fast task, it's helpful to create a lookup dictionary in advance.

````python
artist_to_index = dict(
    zip(
        artists_index["Name"],
        artists_index["Index"]
    )
)
````

### Add ISO Date for Concerts

Next, I'm going to use a utility I wrote some time ago that transforms string dates into ISO format date objects. Then I'll apply the method to every row in the venue_artist_event sheet and generate a new column, "date". Note that this is not writing the new column to the sheet, just adding it to the pandas data frame I'm working from.

````python
import datetime
from dateutil import parser

def str_date_to_iso(month, day, year):
    
    str_date = f"{month} {day} {year}"
    dt = parser.parse(str_date).date().isoformat()
    
    return dt  
````

````python
venue_artist_event["date"] = venue_artist_event.apply(lambda row: str_date_to_iso(row['Month'], row['Day'], row['Year']), axis=1)
````

When I ran the method, it flagged a concert in the data by Bob Wills on "February 29, 1962" at the Farmer's Daughter. There was no February 29 in 1962, so I changed it to February 28.

### Create Table With First and Last Artist Apperances

Next, I'm creating a new data frame with all the artists listed next to their first and last appearances in the concert list, by chronological order.

````python
venue_artist_event_dates = (
    venue_artist_event
    .groupby("Artist_Formula")["date"]
    .agg(["min", "max"])
)
````

````python
venue_artist_event_dates
````

Inspecting this list, I can see a few issues straight away: an artist just called "vocals]', and an artist called "$Not", another one just called "west". These issues may bear further investigation. However, I will postpone data cleanup for now and proceed with organizing the data.

### Create Table With Unique Indexes for All Artists

````python
artist_list_rows = []

next_index = artists_index["Index"].max() +1

for artist in unique_artists:
    
    if artist in artist_to_index:
        idx = artist_to_index[artist]
    else:
        idx = next_index
        next_index += 1

    if artist in venue_artist_event_dates.index:
        first = venue_artist_event_dates.loc[artist, "min"]
        last = venue_artist_event_dates.loc[artist, "max"]
    else:
        first = None
        last = None
        
    artist_list_rows.append({
        "Index": idx,
        "Artist": artist,
        "FirstAppearance": first,
        "LastAppearance": last,
    })

result = pd.DataFrame(artist_list_rows)
````

### Add First and Last Appearance Dates for Venues

As in the case of artists, I would like the first and last appearance dates of any given venue. ### Compare Student and Angelina Coordinates

I want to determine the level of agreement between the student worker's coordinates, derived through manual search, and Angelina's coordinates, derived from an R library query. I generate a script that loads the source data, the student's data, and Angelina's data, and compares them. I set a tolerance of about 22 meters to signal "difference". Only 2 venues have coordinates in both sets within this tolerance. There are seven venues for which only the student found coordinates, and five venues for which only Angelina found coordinates. There are 42 venues for which the coordinates differ by more than 22 meters. 

````python
import pandas as pd
import numpy as np

# ── Load your dataframes ─────────────────────────────────────────────────────
original_df = layer_1_venues  # 974 rows
person1_df  = student_data    # 58 rows, has ID column
person2_df  = angelina_data   # 974 rows, same order as original

# ── Build a comparison frame keyed on original row position ──────────────────
p2_coords = person2_df[["Latitude", "Longitude"]].rename(
    columns={"Latitude": "lat_angelina", "Longitude": "lon_angelina"}
)
p2_coords["_orig_idx"] = p2_coords.index

person1_df["_orig_idx"] = person1_df["ID"].astype(int) - 1
p1_coords = person1_df[["_orig_idx", "LATITUDE", "LONGITUDE"]].rename(
    columns={"LATITUDE": "lat_student", "LONGITUDE": "lon_student"}
)

merged = p1_coords.merge(p2_coords, on="_orig_idx", how="inner")
merged["Venue_Name"] = original_df.loc[merged["_orig_idx"].values, "Venue_Name"].values

# ── Convert to numeric ────────────────────────────────────────────────────────
for col in ["lat_student", "lon_student", "lat_angelina", "lon_angelina"]:
    merged[col] = pd.to_numeric(merged[col], errors="coerce")

# ── Flag differences ──────────────────────────────────────────────────────────
TOLERANCE = 0.0002

merged["lat_diff"]      = (merged["lat_student"] - merged["lat_angelina"]).abs()
merged["lon_diff"]      = (merged["lon_student"] - merged["lon_angelina"]).abs()
merged["both_have_geo"] = merged[["lat_student", "lon_student",
                                   "lat_angelina", "lon_angelina"]].notna().all(axis=1)

# ── Haversine formula ─────────────────────────────────────────────────────────
def haversine_metres(lat1, lon1, lat2, lon2):
    R = 6_371_000
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    dphi    = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi / 2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(dlambda / 2)**2
    return R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

merged["distance_m"] = np.where(
    merged["both_have_geo"],
    haversine_metres(
        merged["lat_student"], merged["lon_student"],
        merged["lat_angelina"], merged["lon_angelina"]
    ),
    np.nan
)

merged["coords_differ"] = merged["both_have_geo"] & (merged["distance_m"] > 22)  # metres threshold

# ── Subset views ──────────────────────────────────────────────────────────────
conflicts = merged[merged["coords_differ"]]
agree     = merged[merged["both_have_geo"] & ~merged["coords_differ"]]
only_p1   = merged[merged["lat_angelina"].isna() & merged["lat_student"].notna()]
only_p2   = merged[merged["lat_student"].isna() & merged["lat_angelina"].notna()]

# ── Display helper ────────────────────────────────────────────────────────────
def print_comparison(df, cols):
    display = df[cols].copy()
    display["Venue_Name"] = display["Venue_Name"].str[:30]
    print(display.to_string(index=False, float_format=lambda x: f"{x:.2f}"))

# ── Summary ───────────────────────────────────────────────────────────────────
print(f"Venues compared:            {len(merged)}")
print(f"  Both have coordinates:    {merged['both_have_geo'].sum()}")
print(f"  Agree (within tolerance): {len(agree)}")
print(f"  DIFFER:                   {len(conflicts)}")
print(f"  Only Student has coords:  {len(only_p1)}")
print(f"  Only Angelina has coords: {len(only_p2)}")

print("\n── Conflicts ───────────────────────────────────────────────────────────")
print_comparison(conflicts, ["_orig_idx", "Venue_Name", "lat_student", "lon_student",
                              "lat_angelina", "lon_angelina", "distance_m"])

print("\n── Only Student has coordinates ────────────────────────────────────────")
print_comparison(only_p1, ["_orig_idx", "Venue_Name", "lat_student", "lon_student"])

print("\n── Only Angelina has coordinates ───────────────────────────────────────")
print_comparison(only_p2, ["_orig_idx", "Venue_Name", "lat_angelina", "lon_angelina"])

print("\n── Agree (within tolerance) ────────────────────────────────────────────")
print_comparison(agree, ["_orig_idx", "Venue_Name",
                          "lat_student", "lon_student", "lat_angelina", "lon_angelina"])

# ── Export ────────────────────────────────────────────────────────────────────
# conflicts.to_csv("conflicts.csv", index=False)
# merged.to_csv("full_comparison.csv", index=False)
````

````python
import pandas as pd
import numpy as np
import os

# ── Load your dataframes ─────────────────────────────────────────────────────
original_df  = layer_1_venues  # 974 rows
mapbox_df    = mapbox_data     # 974 rows, has id column (id = orig_idx + 1)
angelina_df  = angelina_data   # 974 rows, same order as original

# ── Haversine formula ─────────────────────────────────────────────────────────
def haversine_metres(lat1, lon1, lat2, lon2):
    R = 6_371_000
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    dphi    = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi / 2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(dlambda / 2)**2
    return R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

# ── San Antonio centre ────────────────────────────────────────────────────────
SA_LAT = 29.42861139590851
SA_LON = -98.49619652095998

# ── Build comparison frame keyed on original row position ────────────────────
# Mapbox: id = orig_idx + 1
mapbox_df["_orig_idx"] = mapbox_df["id"].astype(int) - 1
mb_coords = mapbox_df[["_orig_idx", "latitude", "longitude"]].rename(
    columns={"latitude": "mapbox_lat", "longitude": "mapbox_long"}
)

# Angelina: already in original order
ang_coords = angelina_df[["Latitude", "Longitude"]].rename(
    columns={"Latitude": "tidyverse_lat", "Longitude": "tidyverse_long"}
)
ang_coords["_orig_idx"] = ang_coords.index

# ── Join on position ──────────────────────────────────────────────────────────
merged = mb_coords.merge(ang_coords, on="_orig_idx", how="inner")
merged["Venue_Name"] = original_df.loc[merged["_orig_idx"].values, "Venue_Name"].values

# ── Convert to numeric ────────────────────────────────────────────────────────
for col in ["mapbox_lat", "mapbox_long", "tidyverse_lat", "tidyverse_long"]:
    merged[col] = pd.to_numeric(merged[col], errors="coerce")

# ── Lat/lon diffs ─────────────────────────────────────────────────────────────
TOLERANCE = 0.0002

merged["lat_diff"] = (merged["mapbox_lat"] - merged["tidyverse_lat"]).abs()
merged["lon_diff"] = (merged["mapbox_long"] - merged["tidyverse_long"]).abs()
merged["both_have_geo"] = merged[["mapbox_lat", "mapbox_long",
                                   "tidyverse_lat", "tidyverse_long"]].notna().all(axis=1)

# ── Distance between the two results ─────────────────────────────────────────
merged["distance"] = np.where(
    merged["both_have_geo"],
    haversine_metres(
        merged["mapbox_lat"], merged["mapbox_long"],
        merged["tidyverse_lat"], merged["tidyverse_long"]
    ),
    np.nan
)

merged["coords_differ"] = merged["both_have_geo"] & (merged["distance"] > 22)  # metres threshold

# ── Distance from central San Antonio ────────────────────────────────────────
merged["mapbox_dist"] = np.where(
    merged[["mapbox_lat", "mapbox_long"]].notna().all(axis=1),
    haversine_metres(merged["mapbox_lat"], merged["mapbox_long"], SA_LAT, SA_LON),
    np.nan
)

merged["tidyverse_dist"] = np.where(
    merged[["tidyverse_lat", "tidyverse_long"]].notna().all(axis=1),
    haversine_metres(merged["tidyverse_lat"], merged["tidyverse_long"], SA_LAT, SA_LON),
    np.nan
)

# ── Subset views ──────────────────────────────────────────────────────────────
conflicts    = merged[merged["coords_differ"]]
agree        = merged[merged["both_have_geo"] & ~merged["coords_differ"]]
only_mapbox  = merged[merged["tidyverse_lat"].isna() & merged["mapbox_lat"].notna()]
only_tidy    = merged[merged["mapbox_lat"].isna() & merged["tidyverse_lat"].notna()]

# ── Summary ───────────────────────────────────────────────────────────────────
print(f"Venues compared:              {len(merged)}")
print(f"  Both have coordinates:      {merged['both_have_geo'].sum()}")
print(f"  Agree (within tolerance):   {len(agree)}")
print(f"  DIFFER:                     {len(conflicts)}")
print(f"  Only Mapbox has coords:     {len(only_mapbox)}")
print(f"  Only Tidyverse has coords:  {len(only_tidy)}")

# Names that script 1 flagged as "mapbox has coords, angelina doesn't"
only_mapbox_names = only_mapbox['Venue_Name'].tolist()
print(only_mapbox_names)

# Are they in all_venues?
for name in only_mapbox_names:
    match = all_venues[all_venues['Venue'] == name]
    print(f"{name!r}: {len(match)} match(es) in all_venues")

def print_comparison(df, cols):
    display = df[cols].copy()
    display["Venue_Name"] = display["Venue_Name"].str[:30]
    print(display.to_string(index=False, float_format=lambda x: f"{x:.2f}"))


print("\n── Conflicts ───────────────────────────────────────────────────────────")
print_comparison(conflicts, ["_orig_idx", "Venue_Name",
                             "mapbox_lat", "mapbox_long",
                             "tidyverse_lat", "tidyverse_long",
                             "distance", "mapbox_dist", "tidyverse_dist"])

print("\n── Only Mapbox has coordinates ─────────────────────────────────────────")
print(only_mapbox[["_orig_idx", "Venue_Name", "mapbox_lat", "mapbox_long"]].to_string(index=False))

print("\n── Only Tidyverse has coordinates ──────────────────────────────────────")
print(only_tidy[["_orig_idx", "Venue_Name", "tidyverse_lat", "tidyverse_long"]].to_string(index=False))

print("\n── Agree (within tolerance) ────────────────────────────────────────────")
print(agree[["_orig_idx", "Venue_Name",
             "mapbox_lat", "mapbox_long",
             "tidyverse_lat", "tidyverse_long"]].to_string(index=False))
# ── Export ────────────────────────────────────────────────────────────────────
# conflicts.to_csv("conflicts_mapbox_vs_tidyverse.csv", index=False)
# merged.to_csv("full_comparison_mapbox_vs_tidyverse.csv", index=False)
````

### Generate comprehensive geolocation table

I've analyzed all three sets of geolocation data, and evaluated their reliability and completeness. The next step is to combine the geolocation data with the comprehensive venues list. The resulting table should have the known data on address and geolocation along with the venue name and identifier.

I've identified the Tidyverse data as the most complete and reliable, so I'll begin by merging that data and prioritize it. I will load the comprehensive venues list from the `VenuesIndex` sheet.

````python
all_venues = pd.read_excel('sounds_data_to_share.xlsx', sheet_name='VenuesIndex')
````

````python
import pandas as pd

venues_with_locations = (
    all_venues[['Index', 'Venue']]
    .merge(
        angelina_df[['Venue_Name', 'Longitude', 'Latitude', 'Address', 'Zip', 'City']],
        left_on='Venue',
        right_on='Venue_Name',
        how='left'
    )
    .drop(columns='Venue_Name')
)

# Format Zip as integer string (no decimals), preserving NaN as empty string
venues_with_locations['Zip'] = (
    venues_with_locations['Zip']
    .apply(lambda x: str(int(float(x))) if pd.notna(x) and str(x).strip() != '' else '')
)
````

Our next source of geolocations, in order by reliability, is Mapbox. According to the analysis there were ten locations identified by Mabpox that Tidyverse did not get locations for. However, Mapbox had a higher tendency to false positives than Tidverse. As a result, I want to build in a quality control check that flags locations too far from San Antonio to be likely matches.

````python
mapbox_data.columns
````

````python
from geopy.distance import geodesic

DOWNTOWN_SA = (29.4241, -98.4936)

# Rename mapbox_data columns to match what we use downstream
mapbox_data = mapbox_data.rename(columns={
    'name': 'Venue_Name',
    'address': 'Address',
    'city': 'City',
    'state': 'State',
    'zip': 'Zip',
    'longitude': 'Longitude',
    'latitude': 'Latitude'
})

# Find rows in venues_with_locations that are missing coordinates
missing_coords_mask = venues_with_locations['Latitude'].isna() | venues_with_locations['Longitude'].isna()

# Also catch corrupted rows like the MacArthur Field one (lat field holding "lat, long")
corrupted_mask = venues_with_locations['Latitude'].astype(str).str.contains(',', na=False)

needs_fill_mask = missing_coords_mask | corrupted_mask
needs_fill = venues_with_locations[needs_fill_mask].copy()

print(f"Venues missing or with corrupted coordinates: {len(needs_fill)}")

filled = []
still_missing = []
too_far = []

for idx, row in needs_fill.iterrows():
    venue_name = row['Venue']
    mapbox_match = mapbox_data[mapbox_data['Venue_Name'] == venue_name]

    if mapbox_match.empty:
        still_missing.append(venue_name)
        continue

    mb_row = mapbox_match.iloc[0]
    lat, lon = mb_row['Latitude'], mb_row['Longitude']

    if pd.isna(lat) or pd.isna(lon):
        still_missing.append(venue_name)
        continue

    distance = geodesic(DOWNTOWN_SA, (lat, lon)).miles
    if distance > 30:
        too_far.append({'Venue': venue_name, 'Distance (miles)': round(distance, 2)})
        continue

    # Update venues_with_locations in place
    venues_with_locations.loc[idx, 'Latitude'] = lat
    venues_with_locations.loc[idx, 'Longitude'] = lon
    venues_with_locations.loc[idx, 'Address'] = mb_row['Address']
    venues_with_locations.loc[idx, 'City'] = mb_row['City']
    venues_with_locations.loc[idx, 'Zip'] = (
        str(int(mb_row['Zip'])) if pd.notna(mb_row['Zip']) and str(mb_row['Zip']).strip() != '' else ''
    )
    filled.append(venue_name)

print(f"\nFilled coordinates for {len(filled)} venues: {filled}")

if too_far:
    print("\nVenues with mapbox coords more than 30 miles from downtown SA (not filled):")
    for v in too_far:
        print(f"  {v['Venue']}: {v['Distance (miles)']} miles")

if still_missing:
    print(f"\nStill missing coordinates ({len(still_missing)}): {still_missing}")
````

This script went through the mapbox data, looked for any names matching a name in our comprehensive venues list, and checked if there was already a set of valid coordinates there. If there was not, it took the coordinates from the data and put them in the list. 

It found 19 venues with either missing or bad coordinates and added or replaced that data. 

It also ran a check to see if a particular venue was further than 30 miles from San Antonio, and if it was, it didn't add the coordinates. Some of these are clearly errors. However, Twin Palms, Twin Mountain Inn, and Seven Oaks might be correctly identifed and should be investigated.

After this, we still have 1998 venues with no coordinates.
